```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='x vs x')
ax.plot(x,x**2, label='x vs x**2')

ax.legend(loc='lower left')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='x vs x')
ax.plot(x,x**2, label='x vs x**2')

ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple',label='x vs x') #RGB HEX Code
ax.plot(x,x+1,color='#e6c017',label='x vs x+1')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple', lw=10) #RGB HEX Code
ax.plot(x,x+1,color='#e6c017', linewidth=0.5)

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple', ls='--') #RGB HEX Code
ax.plot(x,x+1,color='#e6c017', linestyle='-.', lw=4)

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='purple', marker='+', markersize=20) #RGB HEX Code
ax.plot(x,x+1,color='#e6c017', marker='o')

```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a **4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
fig, axes = plt.subplots(2,2)

axes[0][0].plot(a,b)
axes[1][0].plot(x,y)
```

```python
fig, axes = plt.subplots(3,1)

axes[0].plot(a,b)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2,2)

axes[0][0].plot(a,b)
axes[1][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][1].plot(x,y)

fig.subplots_adjust(wspace=0.9, hspace=0.4)
```

```python
fig, axes = plt.subplots(2,2)

axes[0][0].plot(a,b)

axes[1][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][1].plot(x,y)
axes[1][1].set_title('TITLE')
axes[1][1].set_ylabel('Y LABEL 1,1')
axes[1][1].set_xlim(2,6)

fig.suptitle('Figure Level Title')
plt.tight_layout()

fig.savefig('myfirstsubplots.png', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(2,2, figsize=(4,10), dpi=200)

axes[0][0].plot(a,b)

axes[1][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][1].plot(x,y)
axes[1][1].set_title('TITLE')
axes[1][1].set_ylabel('Y LABEL 1,1')
axes[1][1].set_xlim(2,6)

fig.suptitle('Figure Level Title')
plt.tight_layout()

# fig.savefig('myfirstsubplots.png', bbox_inches='tight')
```
